```python
import pandas as pd

PopulationData_path = '../../data/week 1/population.csv'
population_df = pd.read_csv(PopulationData_path)
population_df.head()
```

```python
# remove some useless columns from the 'temp_df' data frame.
useless_col = ["Domain Code", "Domain", "Element Code", "Element","Item Code","Item", "Year Code"]
population_df.drop(labels=useless_col, inplace=True, axis=1)
population_df.head()
```

```python
YieldData_path = '../../data/week 1/yield.csv'
yield_df = pd.read_csv(YieldData_path)
yield_df.head()
```

```python
useless_col = ["Domain Code", "Domain", "Element Code", "Element","Item Code (FAO)","Item", "Year Code"]
yield_df.drop(labels=useless_col, inplace=True, axis=1)
yield_df.head()
```

```python
AreaharvestedData_path = '../../data/week 1/harvested.csv'
harvested_df = pd.read_csv(AreaharvestedData_path)
harvested_df.head()
```

```python
useless_col = ["Domain Code", "Domain", "Element Code", "Element","Item Code (FAO)","Item", "Year Code"]
harvested_df.drop(labels=useless_col, inplace=True, axis=1)
harvested_df.head()
```

```python
ProductionData_path = '../../data/week 1/production.csv'
production_df = pd.read_csv(ProductionData_path)
production_df.head()
```

```python
useless_col = ["Domain Code", "Domain", "Element Code", "Element","Item Code (FAO)","Item", "Year Code"]
production_df.drop(labels=useless_col, inplace=True, axis=1)
production_df.head()
```

```python
Population_yield=pd.merge(population_df, yield_df, how="inner", on=["Area Code (FAO)","Year"])

Population_yield[0:13]
```

```python
# removing useless columns from 'TempWithWheat' dataFrame.
Population_yield.drop(inplace=True, labels=['Area_y', 'Area_x'], axis=1)

# show only 2 rows randomly.
Population_yield.sample(2)
```

```python
Population_yield_harvested=pd.merge(Population_yield, harvested_df, how="inner", on=["Area Code (FAO)","Year"])
Population_yield_harvested = Population_yield_harvested.rename(columns={'Value_x': 'population','Value_y':'yield','Value':'harvested'})
Population_yield_harvested[0:13]
```

```python
# removing useless columns from 'TempWithWheat' dataFrame.
Population_yield_harvested.drop(inplace=True, labels=['Area'], axis=1)

# show only 2 rows randomly.
Population_yield_harvested.sample(2)
```

```python
Population_yield_harvested_production=pd.merge(Population_yield_harvested, production_df, how="inner", on=["Area Code (FAO)","Year"])
Population_yield_harvested_production[0:13]
```

```python
# removing useless columns from 'TempWithWheat' dataFrame.
Population_yield_harvested_production.drop(inplace=True, labels=['Area'], axis=1)

# show only 2 rows randomly.
Population_yield_harvested_production.sample(2)
```

```python
# now it's time to normalize our data, we use MinMaxScaler for this task
# (this is like what we did in issue 3)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
Population_yield_harvested_production[['population', 'yield', 'harvested', 'Value']] = scaler.fit_transform(Population_yield_harvested_production[['population', 'yield', 'harvested', 'Value']])

# as there are too many decimal digits we are going to use only 4 decimals.
Population_yield_harvested_production['population'] = Population_yield_harvested_production['population'].round(decimals=4)
Population_yield_harvested_production['yield'] = Population_yield_harvested_production['yield'].round(decimals=4)
Population_yield_harvested_production['harvested'] = Population_yield_harvested_production['harvested'].round(decimals=4)
Population_yield_harvested_production['Value']  = Population_yield_harvested_production['Value'].round(decimals=4)
Population_yield_harvested_production.sample(2)
```

```python
Population_yield_harvested_production.to_csv('../../data/week 2/merge.csv', index=False)
```
